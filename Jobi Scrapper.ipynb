```python
result = df
for i in range(20):
    df1 =pd.read_json("https://www.jobi.tn/api/joboffers/sponsored?limit=50")
    df1.head()
    result = pd.concat([result, df1], sort=False)
```

```python
saved_result = result
result.head()
```

```python
result['salary']=(result.salary_min+result.salary_max)/2
```

```python
result[['salary_max', 'salary_max', 'salary']]
```

```python
result['salary_min'].fillna(df['salary_max'],inplace=True)
result['salary_max'].fillna(df['salary_min'],inplace=True)

result['salary']=(result.salary_min+result.salary_max)/2
salary_mean = result.salary.mean()
result['salary'].fillna(salary_mean,inplace=True)
```
