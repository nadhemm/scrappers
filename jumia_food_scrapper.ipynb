```python
import time

from selenium import webdriver
from selenium.webdriver.common.keys import Keys

driver = webdriver.Firefox()
driver.implicitly_wait(5)
```

```python
restaurants={}
all_restaurants = []
total_regions = 111
links={}
```

```python
for x in range(total_regions):
    try:
        driver.get("https://food.jumia.com.tn")
#         time.sleep(1)
        a = driver.find_element_by_xpath("//input[@type='search']")
        a.click()
        a.send_keys(Keys.ARROW_DOWN)
        for i in range(50):
            a.send_keys(Keys.BACKSPACE)
        for y in range(x+1):
            a.send_keys(Keys.ARROW_DOWN)    
        a.send_keys(Keys.ENTER)
        a.send_keys(Keys.ENTER)
        btn = driver.find_element_by_xpath("//button[@type='submit']")
        btn.click()
#         time.sleep(1)
        
        region = driver.find_element_by_xpath("//div[@class='change-location']").text
        print("#", x, "processing:", region)
        for i in range(5):
            time.sleep(1)
            driver.execute_script("window.scrollTo(0,10000000)")
        restaurants[region]=[]
        a = driver.find_elements_by_tag_name("a")
        i =0
        for el  in a[12:-2]:
            i+=1
#             if el.get_attribute('href') in links:
#                 print(el.get_attribute('href'))
            links[el.get_attribute('href')]=region
    except Exception as e:
        print("Exception in region", x ,'details',e)
        pass
```

```python
len(links)
```

```python
for link in links:
    try:
        print("*"*30)
        print(link)
        driver.get(link)
#                 time.sleep(1)
        name = driver.find_element_by_xpath("//h1[@class='vendor-header-title']").text
        print(name)
        res = driver.find_element_by_xpath("//*[contains(text(),'informations')]")
        res.click()
#                 time.sleep(.5)
        address = driver.find_elements_by_xpath("//div[@class='pbm']")
        adr = address[-1].text.replace("\n", " ")
        print(adr)
        restaurants[region].append({"name":name, "address":adr})        
        all_restaurants.append({"name":name, "address":adr})        
    except Exception as e:
        print("error occurred in restaurant:", link, "details:", e)
```

```python
print(restaurants)
print(all_restaurants)
```

```python
import json
with open('links.json', 'w') as outfile:
    json.dump(links, outfile, ensure_ascii=False)
```

```python
import json
with open('all_restaurants.json', 'w') as outfile:
    json.dump(all_restaurants, outfile, ensure_ascii=False)
    
with open('restaurants_per_region.json', 'w') as outfile:
    json.dump(restaurants, outfile, ensure_ascii=False)
    
    
```
