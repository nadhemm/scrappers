# Scrapper for *www.talents.tn*

```python
import pandas as pd
import requests
import json
from bs4 import BeautifulSoup
```

```python
offers=[]
titles=[]
text=""
for page in range(1, 77):
    print("processing page", page)
    link = "https://www.talents.tn/rechercheroffres/?keywords=a&countrycode=FR&Page="+str(page)
    req = requests.get(link)
    soup = BeautifulSoup(req.content,'lxml')
    #         res= soup.find_all("div",{"class": "li-item-list"})
    #         scrap_res(res)
    res= soup.find_all("div",{"class": "js-clickable"})
    for a in res:
        address = a.ul.li.get_text()
        title = a.h3.get_text()
        titles.append(title)
        offers.append({'title':title, 'address':address})
        text+=' '+title
```

```python
offers
```

```python
import operator
occ={}
occurrences={}
BLACK_LIST = ["h/f","que"]
for sub in text.lower().split():
    if len(sub.strip())>2 and sub.strip() not in BLACK_LIST:
        occ[sub.strip()]=text.lower().count(sub.strip())
occ = sorted(occ.items(), key=operator.itemgetter(1))
for w in occ:
    occurrences[w[0]]=w[1]
    print(w[0] ,w[1])
    
```

```python
for o in occurrences:
    print(occurrences[o], o)
```

```python
with open('occurrences.json', 'w') as f:  # writing JSON object
    json.dump(occurrences, f, ensure_ascii=False)
```

```python
with open('titles.json', 'w') as f:  # writing JSON object
    json.dump({'titles':titles}, f, ensure_ascii=False)
```

```python
with open('offers.json', 'w') as f:  # writing JSON object
    json.dump(offers, f, ensure_ascii=False)
```
